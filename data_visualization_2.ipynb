```python
# Пример визуализации с помощью Matplot

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# print(plt.style.available)

x = np.linspace(0, 10, 100) # создание массива значений
y = np.sin(x)

sns.set_style("ticks")
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['georgia']
plt.figure(figsize=(6, 3.5))

plt.plot(x, y, color='black', linewidth=2, alpha=0.7)
plt.fill_between(x, y, color='green', alpha=0.2)
plt.xlabel('$x$') # использование LaTeX
plt.ylabel('$sin(x)$')
plt.ylim(-1.2, 1.2)
plt.title('График функции синуса')
plt.show()
```

```python
# Базовые операции с числами в Python

a = 10
b = 5
sum = a + b
difference = a - b
product = a * b
quotient = a / b

print("Sum:", sum)
print("Difference:", difference)
print("Product:", product)
print("Quotient:", quotient)
```

```python
# Создание простого графика с помощью Matplotlib

x = np.linspace(0, 10, 50)
y = x**2

sns.set_style("ticks")
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['georgia']
plt.figure(figsize=(6, 3.5))

plt.plot(x, y, color='black', linewidth=2, alpha=0.7)
plt.fill_between(x, y, color='green', alpha=0.2)
plt.xlabel("$x$")
plt.ylabel("$x^2$") # использование LaTeX для верхнего индекса
plt.title("График функции $y = x^2$")
plt.show()
```

```python
# Работа со списками

cities = ["Минск", "Санкт-Петербург", "Гродно", "Сочи", "Москва"]

print("Третий город в списке:", cities[2])

cities.append("Гомель")
del cities[1] # оператор del не требует использования скобок, как функции, хотя скобки можно добавить

print("Обновленный список городов:", cities)
```

```python
# Визуализация данных

months = ["Январь", "Февраль", "Март", "Апрель", "Май", "Июнь", 
          "Июль", "Август", "Сентябрь", "Октябрь", "Ноябрь", "Декабрь"]

temperatures = [-5, -3, 5, 12, 18, 22, 25, 24, 18, 10, 2, -2]

sns.set_style("ticks")
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['georgia']
plt.figure(figsize=(10, 3.3))

plt.plot(months, temperatures, color='black', linewidth=2, alpha=0.7)
plt.fill_between(months, temperatures, color='green', alpha=0.2)
plt.ylim(-7, 27)
plt.xlabel("Месяцы")
plt.ylabel("Температура, °C")
plt.title("Средняя температура в течение года")
plt.grid(color='gray', linestyle='-.', linewidth=0.5, alpha=0.5)
plt.show()
```

```python
# Дополнительные задания

import matplotlib.pyplot as plt

# Данные для столбчатой диаграммы
categories = ['Категория A', 'Категория B', 'Категория C', 'Категория D']
values = [4, 7, 1, 8]

# Создание столбчатой диаграммы
plt.figure(figsize=(6, 4))
plt.bar(categories, values, color='green', alpha=0.6)
plt.xlabel('Категории')
plt.ylabel('Значения')
plt.title('Пример столбчатой диаграммы')
plt.grid(axis='y', linewidth=0.5, alpha=0.5)
plt.show()
 
```
